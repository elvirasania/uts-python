```python
# HASIL KLASIFIKASI

s1 = probabilitas_lank_x_s1 * probabilitas_lank_y_s1 * probabilitas_lank_z_s1
s2 = probabilitas_lank_x_s2 * probabilitas_lank_y_s2 * probabilitas_lank_z_s2
s3 = probabilitas_lank_x_s3 * probabilitas_lank_y_s3 * probabilitas_lank_z_s3
s4 = probabilitas_lank_x_s4 * probabilitas_lank_y_s4 * probabilitas_lank_z_s4

def identitySub(x, y, z, t):
  if x > y and x > z and x > t:
    y = 'SUBYEK 1'
  elif y > x and y > z and y > t:
    y = 'SUBYEK 2'
  elif z > x and z > y and z > t:
    y = 'SUBYEK 3'
  elif t > x and t > y and t > z:
    y = 'SUBYEK 4'
  
  return y

print('HASIL KLASIFIKASI')
print('SUBYEK 1 : ', format(float(s1), '.20f'))
print('SUBYEK 2 : ', format(float(s2), '.20f'))
print('SUBYEK 3 : ', format(float(s3), '.20f'))
print('SUBYEK 4 : ', format(float(s4), '.20f'))
print('\n')
print('Berdasarkan hasil klasifikasi maka yang terbesar adalah hasil \nakhirnya yang akan menjadi prediksi selanjutnya')

print('RESULT       : ', identitySub(s1, s2, s3, s4))
```
